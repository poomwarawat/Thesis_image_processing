```python
from skimage import io
from skimage import img_as_float
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_img = io.imread("python/test_img.jpg")
plt.imshow(my_img)
plt.title("Original Image")
print(my_img.min(), my_img.max())
```

```python
my_float_img = img_as_float(my_img)
plt.imshow(my_float_img)
plt.title("Float Image")
print(my_float_img.min(), my_float_img.max())
```

```python
random_img = np.random.random([500, 500])
plt.imshow(random_img)
print(random_img.min(), random_img.max())
```
