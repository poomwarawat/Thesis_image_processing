```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("python/BSE_Google_noisy.jpg", 0)
```

```python
img.shape
```

```python
kernal = np.ones((5, 5), np.float32)/25
```

```python
filt_2D = cv2.filter2D(img, -1, kernal)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(14, 14))

ax = axes.ravel()

ax[0].imshow(img)
ax[0].set_title("Original Image")

ax[1].imshow(filt_2D)
ax[1].set_title("Filter Image")

for x in ax:
    x.axis("off")
    
plt.tight_layout()
plt.show()
```

```python
blur = cv2.blur(img, (3, 3))
```

```python
gaussian = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(14, 14))

ax = axes.ravel()

ax[0].imshow(blur)
ax[0].set_title("Blur Image")

ax[1].imshow(gaussian)
ax[1].set_title("Gaussian Blur Image")

for x in ax:
    x.axis("off")
    
plt.tight_layout()
plt.show()
```

```python
plt.imshow(bilateral)
plt.axis("off")
plt.title("Bilateral Image")
```

# Threshold

```python
alloy = cv2.imread("python/Alloy1.jpg", 0)
plt.imshow(alloy, cmap="gray")
```

```python
eq_img = cv2.equalizeHist(alloy)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(4, 4))

ax = axes.ravel()

ax[0].hist(alloy.flat, bins=100, range=(0, 255))
ax[0].set_title("Alloy Hist")

ax[1].hist(eq_img.flat, bins=100, range=(0, 255))
ax[1].set_title("EQ Alloy Hist")
    
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(8, 8))

ax = axes.ravel()

ax[0].imshow(alloy, cmap="gray")
ax[0].set_title("Alloy Image")

ax[1].imshow(eq_img, cmap="gray")
ax[1].set_title("EQ Image")
    
plt.tight_layout()
plt.show()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
cl_img = clahe.apply(alloy)
```

```python
plt.imshow(cl_img, cmap="gray")
plt.title("Clahe Image")
```

```python
plt.hist(cl_img.flat, bins=100, range=(0, 255))
plt.title("Clahe Hist")
```

```python
ret, thresh = cv2.threshold(cl_img, 190, 150, cv2.THRESH_BINARY)
ret2, thresh2 = cv2.threshold(cl_img, 190, 150, cv2.THRESH_BINARY_INV)
ret3, thresh3 = cv2.threshold(cl_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(10, 10))

ax = axes.ravel()

ax[0].imshow(thresh, cmap="gray")
ax[0].set_title("Thresh Image")

ax[1].imshow(thresh2, cmap="gray")
ax[1].set_title("Thresh2 Image")

ax[2].imshow(thresh3, cmap="gray")
ax[2].set_title("Thresh3 Image")
    
plt.tight_layout()
plt.show()
```

```python
plt.hist(img.flat, bins=100, range=(0, 255))
plt.title("BSE Noisy Histogram")
```

```python
ret4, thresh4 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.imshow(thresh4, cmap="gray")
```

```python
kernal2 = np.ones((3, 3), np.uint8)
```

```python
erosion = cv2.erode(thresh4, kernal2, iterations=1)
dilation = cv2.dilate(erosion, kernal2, iterations=1)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(10, 10))

ax = axes.ravel()

ax[0].imshow(erosion, cmap="gray")
ax[0].set_title("Erosion Image")

ax[1].imshow(dilation, cmap="gray")
ax[1].set_title("Dilation Image")
    
plt.tight_layout()
plt.show()
```

```python
opening = cv2.morphologyEx(thresh4, cv2.MORPH_OPEN, kernal2)
```

```python
plt.imshow(opening, cmap="gray")
```
