```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, img_as_float
from scipy import ndimage as nd

%matplotlib inline
```

```python
def gaussian_kernel(size, size_y=None):
    size = int(size)
    if not size_y:
        size_y = size
    else:
        size_y = int(size_y)
    
    x, y = np.mgrid[-size:size+1, -size_y:size_y+1]
    g = np.exp(-(x**2/float(size) + y ** 2/float(size_y)))
    return g/g.sum()
```

```python
gaussian_kernel_array = gaussian_kernel(5)
```

```python
plt.imshow(gaussian_kernel_array, cmap=plt.get_cmap('jet'), interpolation='nearest')
plt.colorbar()
plt.show()
```

```python
img = img_as_float(io.imread("python/wbc.jpg"))
```

```python
gaussian_img = nd.gaussian_filter(img, sigma=3)
median_img = nd.median_filter(img, size=3)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(14, 14))

ax = axes.ravel()

ax[0].imshow(img)
ax[0].set_title("Original Image")

ax[1].imshow(gaussian_img)
ax[1].set_title("Gaussian Image")

ax[2].imshow(median_img)
ax[2].set_title("Median Image")

for x in ax:
    x.axis("off")
    
plt.tight_layout()
plt.show()
```

```python
from skimage.restoration import denoise_nl_means, estimate_sigma
```

```python
sigma_est = np.mean(estimate_sigma(img, multichannel=True))
nlm = denoise_nl_means(img, h=1.15 * sigma_est, fast_mode=False, 
                           patch_size=5, patch_distance=3, multichannel=True)
```

```python
plt.imshow(nlm)
plt.title("Denoise image")
plt.axis("off")
```
